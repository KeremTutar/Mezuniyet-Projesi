```python
from os import listdir
import cv2 as cv 
import numpy as np 
from os.path import isfile , join 
import matplotlib.pyplot as plt 
```

```python
mypath = 'e-scooter rider dataset'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = np.empty(len(onlyfiles), dtype=object)
images_rgb = np.empty(len(onlyfiles), dtype=object)
for n in range(0, len(onlyfiles)):
  images[n] = cv.imread( join(mypath,onlyfiles[n]) )
  images_rgb[n] = cv.cvtColor(images[n] , cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (10,10))
for i in range(1,7) :
    plt.subplot(2,3,i)
    plt.imshow(images_rgb[i])
```

```python
noisy_img = np.empty(10, dtype=object)
for j in range(0,len(onlyfiles)) :
    noise = np.zeros(images_rgb[j].shape, np.uint8)
    mean = 500
    stddev = 900
    cv.randn(noise, mean, stddev)
    noisy_img[j] = cv.add(images_rgb[j], noise)

plt.figure(figsize = (10,10))
for k in range(1,7) :
    plt.subplot(2,3,k)
    plt.imshow(noisy_img[k])
```
